```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# sns built in datasets:
sns.get_dataset_names()
```

```python
# sns load built in datasets:
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
tips.head()
```

# Scatterplot

```python
sns.scatterplot(x='tip', y='total_bill', data=tips)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day')
```

```python
sns.scatterplot(x='tip', y='total_bill', 
                data=tips, 
                hue='day', 
                size="size")
```

```python
sns.scatterplot(x='tip', y='total_bill',
                data=tips,
                hue='day',
                size='size',
                palette='coolwarm')
```

# Histogram:

```python
sns.histplot(tips['tip'], kde=True, bins=15)
# kde = True shows the line
```

```python
sns.histplot(tips['tip'], kde=True, bins=30, color='red')
```

# Distribution Plot:

```python
sns.displot(tips['tip'], kde=True, bins=15, color='green')
```

# Bar Plots:

```python
sns.barplot(x="sex", y="tip", data=tips, palette="Blues_d")
```

# BoxPlot:

```python
sns.boxplot(x='day', y='total_bill', 
            data=tips, 
            hue='sex', 
            palette='rainbow')
```

# Strip Plot

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu', dodge=True)
```

# Joint plots: 
Scatter plot + Distribution Plot

```python
sns.jointplot(x='tip', y='total_bill', data=tips)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', fill=True, cmap='viridis')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='resid')
```

# Heatmap

```python
df = pd.read_csv('titanic-data.csv')
df.head()
```

```python
df2 = df
df2.head()
```

```python
df2.drop(['Name', 'Cabin', 'Embarked'], axis=1, inplace=True)
```

```python
df3 = df2
df3.Sex = df2.Sex.map(lambda x: 1 if x=='male' else 0)
```

```python
df4 = df3[['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch']]
sns.heatmap(df4.corr())
```
